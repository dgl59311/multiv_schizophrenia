### Multivariate distance correlation and t test
### Multiverse 
Run t test between the EEG features and calculate fx size using unbiased distance correlations

```r
# dario.gordillolopez@epfl.ch
library('energy')
library('dplyr')
```

```r
# set working directory
setwd("")
# csv data directory
eeg_data_dir = ""
```

```r
# Find EEG files in .csv 
temp <- list.files(path = eeg_data_dir ,pattern = "*.csv")
```

```r
# allocate memory

# multivariate distance correlations
# for healthy controls
pval_dcor_test_con <- matrix(0,length(temp),length(temp))
fx_dcor_test_con <- matrix(0,length(temp),length(temp))
# for older adults
pval_dcor_test_pat <- matrix(0,length(temp),length(temp))
fx_dcor_test_pat <- matrix(0,length(temp),length(temp))
```

```r
# feature 1
i<-1
# feature 2
j<-1

# load EEG data feature 1 
data_EEG_1 <- read.csv(paste0(eeg_data_dir,temp[i]))
# data for controls
con_eeg_1 <- filter(data_EEG_1,Group=='Controls')[,2:dim(data_EEG_1)[2]]  
# data for patients
pat_eeg_1 <- filter(data_EEG_1,Group=='SZPatients')[,2:dim(data_EEG_1)[2]]  

# load EEG data feature 2
data_EEG_2 <- read.csv(paste0(eeg_data_dir,temp[j]))
# data for controls
con_eeg_2 <- filter(data_EEG_2,Group=='Controls')[,2:dim(data_EEG_2)[2]]  
# data for patients
pat_eeg_2 <- filter(data_EEG_2,Group=='SZPatients')[,2:dim(data_EEG_2)[2]]  

multivar_energy_test_y <- dcorT.test(con_eeg_1,con_eeg_2)
multivar_energy_test_y
```

```r
for (i in 1:length(temp)){
  
  message('Processing feature ', i, ' of ', length(temp))
    
  # load EEG data feature 1 
  data_EEG_1 <- read.csv(paste0(eeg_data_dir,temp[i]))
  # data for controls
  con_eeg_1 <- filter(data_EEG_1,Group=='Controls')[,2:dim(data_EEG_1)[2]]  
  # data for patients
  pat_eeg_1 <- filter(data_EEG_1,Group=='SZPatients')[,2:dim(data_EEG_1)[2]]  
    
  for (j in i:length(temp)){

    # load EEG data feature 2
    data_EEG_2 <- read.csv(paste0(eeg_data_dir,temp[j]))
    # data for controls
    con_eeg_2 <- filter(data_EEG_2,Group=='Controls')[,2:dim(data_EEG_2)[2]]  
    # data for patients
    pat_eeg_2 <- filter(data_EEG_2,Group=='SZPatients')[,2:dim(data_EEG_2)[2]]  
    
    # multivariate distance correlation test between EEG features
    # for controls
    multivar_ttest_con <- dcorT.test(con_eeg_1,con_eeg_2)
    fx_dcor_test_con[i,j] <- multivar_ttest_con$estimate
    pval_dcor_test_con[i,j] <- multivar_ttest_con$p.value
    
    # for patients
    multivar_ttest_pat <- dcorT.test(pat_eeg_1,pat_eeg_2)
    fx_dcor_test_pat[i,j] <- multivar_ttest_pat$estimate
    pval_dcor_test_pat[i,j] <- multivar_ttest_pat$p.value      

  }
}
```

```r
# Results for controls
# write csv with fx magnitude
temp_ix <- gsub(".csv","",temp)
cols_fx <- data.frame()
cols_fx <- as.data.frame(fx_dcor_test_con,row.names = temp_ix)
colnames(cols_fx) <- temp_ix
write.csv(cols_fx,"")

# write csv with pvalues
cols_pval <- data.frame()
cols_pval <- as.data.frame(pval_dcor_test_con,row.names = temp_ix)
colnames(cols_pval) <- temp_ix
write.csv(cols_pval,"")


# Results for patients
# write csv with fx magnitude
cols_fx <- data.frame()
cols_fx <- as.data.frame(fx_dcor_test_pat,row.names = temp_ix)
colnames(cols_fx) <- temp_ix
write.csv(cols_fx,"")

# write csv with pvalues
cols_pval <- data.frame()
cols_pval <- as.data.frame(pval_dcor_test_pat,row.names = temp_ix)
colnames(cols_pval) <- temp_ix
write.csv(cols_pval,"")
```
